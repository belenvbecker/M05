## Analisis de data


- pk1 : porcentaje de personas que entran en cola pero no son atendidos

```python
pk1 = len(callCenter[(callCenter["q_start"] == "0:00:00") & (callCenter["q_exit"] == "0:00:00")])/len(callCenter)
```

- pk2: Promedio de llamadas atendidas por agente

```python
num_agents = len(callCenter[callCenter['outcome'] == "AGENT"])
num_servers = len(callCenter['server'][callCenter['server'] != "NO_SERVER" ].unique())

print(f"El promedio de la llamadas por agente es: {num_agents/num_servers}")
```

```python
cant_call_agente = callCenter[callCenter['outcome'] == "AGENT"][['call_id','server']].groupby('server').count().sort_values(by='call_id',ascending=False)
```

```python
cant_call_agente['Porcentaje'] =  (cant_call_agente['call_id']/num_agents)*100

print("El porcentaje de llamadas por agente es: ")

cant_call_agente.head(5)
```
